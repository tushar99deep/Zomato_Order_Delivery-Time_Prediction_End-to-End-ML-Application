```python
from math import radians, sin, cos, sqrt, atan2
import pandas as pd

def calculate_distance(row):
    # Earth radius in kilometers
    R = 6371.0

    # Convert coordinates to radians
    restaurant_lat = radians(row['Restaurant_latitude'])
    restaurant_lon = radians(row['Restaurant_longitude'])
    delivery_lat = radians(row['Delivery_location_latitude'])
    delivery_lon = radians(row['Delivery_location_longitude'])

    # Calculate the differences between latitudes and longitudes
    dlat = delivery_lat - restaurant_lat
    dlon = delivery_lon - restaurant_lon

    # Apply the haversine formula
    a = sin(dlat/2)**2 + cos(restaurant_lat) * cos(delivery_lat) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))

    # Calculate the distance in kilometers
    distance = R * c
    return distance

# Apply the distance calculation to each row and create a new column 'Distance'
df['Distance'] = df.apply(calculate_distance, axis=1)


'''calculate_distance function takes a row as an argument, and the coordinates are accessed using 
the row indexing (row['Restaurant_latitude'], row['Restaurant_longitude'], etc.). 
The function calculates the distance and returns it.

The .apply() function is then used on the DataFrame to apply the calculate_distance function to each row, 
resulting in a new column named 'Distance' that contains the calculated distances.'''

'''the coordinates in your DataFrame are in decimal degrees (not in radians or any other format) and that the 
column names in the function call match the column names in your DataFrame exactly. 
The resulting distance will be in kilometers, assuming the Earth's radius of 6371 kilometers.'''
```

```python
def value_count_norm(df, feature):
    value_count = df[feature].value_counts()
    value_count_norm_1 = df[feature].value_counts(normalize=True) * 100
    value_count_norm_1_concat = pd.concat([value_count, value_count_norm_1], axis=1)
    value_count_norm_1_concat.columns = ['Count', 'Frequency %']
    return value_count_norm_1_concat
```

```python
def create_pie_chart(df, feature):
    n = len(value_count_norm(df, feature).index)
    colors = sns.color_palette('pastel')[0:n]
    plt.pie(value_count_norm(df, feature).iloc[:, 1],
            colors = colors, autopct='%1.2f%%', 
            labels=value_count_norm(df, feature).index)
    plt.title('{}'.format(feature))
    return plt.show()
```

```python
def create_hist_plot(df, feature):
    figure(figsize=(20, 8))
    sns.histplot(df[feature], kde=True)
    plt.title('{}'.format(feature))
    return plt.show()
```

```python
def get_report(df, feature):
    match feature:
        case 'Weather_conditions'| 'Road_traffic_density'| 'Vehicle_condition'| 'Type_of_order'| 'Type_of_vehicle'| 'multiple_deliveries'| 'Festival'| 'City':
            get_info(df, feature)
            print('*'*50)
            create_pie_chart(df, feature)
        case 'Delivery_person_Age'| 'Delivery_person_Ratings':
            get_info(df, feature)
            print('*'*50)
            create_hist_plot(df, feature)
        case _:
            print("Error")
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.base import BaseEstimator, TransformerMixin

class LabelEncoderTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.label_encoders = {}
    
    def fit(self, X, y=None):
        for column in X.columns:
            self.label_encoders[column] = LabelEncoder()
            self.label_encoders[column].fit(X[column])
        return self
    
    def transform(self, X, y=None):
        transformed_X = X.copy()
        for column in transformed_X.columns:
            transformed_X[column] = self.label_encoders[column].transform(transformed_X[column])
        return transformed_X

def get_data_transformation_object():
    # Defining categorical features for label encoding
    categorical_features = [
        'Weather_conditions',
        'Road_traffic_density',
        'Type_of_order',
        'Type_of_vehicle',
        'Festival',
        'City'
    ]

    numerical_features = [
        'Delivery_person_Age',
        'Vehicle_condition',
        'multiple_deliveries',
        'Distance in KM',
    ]

    # Numerical Pipeline
    num_pipeline = Pipeline(
        steps=[
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ]
    )

    # Categorical Pipeline
    cat_pipeline = Pipeline(
        steps=[
            ('encoder', LabelEncoderTransformer())
        ]
    )

    preprocessor = ColumnTransformer([
        ('num_pipeline', num_pipeline, numerical_features),
        ('cat_pipeline', cat_pipeline, categorical_features)
    ])

    return preprocessor


abc = get_data_transformation_object()
abc.fit(X_train)

```
